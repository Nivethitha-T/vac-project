```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Label'],color='red',bins=4)
plt.title('Histogram')
plt.show()
```

```python
plt.scatter(df['Label'],df['Attempted Category'],color='red')
plt.title('Scatter Plot')
plt.xlabel('Label')
plt.ylabel('Attempted Category')
plt.show()
```

```python
plt.plot(df['Label'],df['Attempted Category'],color='red')
plt.title('Line Plot')
plt.xlabel('Label')
plt.ylabel('Attempted Category')
plt.show()
```

```python
plt.bar(df['Label'],df['Attempted Category'],color='red')
plt.title('Bar Plot')
plt.xlabel('Label')
plt.ylabel('Attempted Category')
plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.title('Sub Plot')
plt.subplot(2,2,1)
plt.title('Histogram')
plt.hist(df['Label'],color='red',bins=4)
plt.xticks(rotation=45,ha='right')
plt.subplot(2,2,2)
plt.scatter(df['Label'],df['Attempted Category'],color='red')
plt.title('Scatter Plot')
plt.xlabel('Label')
plt.ylabel('Attempted Category')
plt.xticks(rotation=45,ha='right')
plt.subplot(2,2,3)
plt.plot(df['Label'],df['Attempted Category'],color='red')
plt.title('Line Plot')
plt.xlabel('Label')
plt.ylabel('Attempted Category')
plt.xticks(rotation=45,ha='right')
plt.subplot(2,2,4)
plt.bar(df['Label'],df['Attempted Category'],color='red')
plt.title('Bar Plot')
plt.xlabel('Label')
plt.ylabel('Attempted Category')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.savefig('plot.png')
plt.show()
```
